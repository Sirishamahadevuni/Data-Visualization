A Histogram is a graphical display of data using bars of different heights.in a histogram,each bar group numbers into ranges.Taller bars show that more data falls  in that range.A histogram displays the shape and spread of continue sample data 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
marks=np.random.randint(0,100,(50))
marks
```

```python
plt.hist(marks)
plt.show()
```

```python
bims=np.arange(0,100,5)
plt.figure(figsize=(6,6))
plt.hist(marks,bins=bims,color='green')
plt.xticks(np.arange(0,100,5),color='orange')
plt.show()
```

```python
bims=np.arange(0,100,5)
plt.figure(figsize=(6,6))
plt.hist(marks,bins=bims,color='green',orientation='horizontal')
plt.yticks(np.arange(0,100,5),color='orange')
plt.show()
```

```python
bims=np.arange(0,100,5)
plt.figure(figsize=(6,6))
plt.hist(marks,bins=bims,color='green',rwidth=0.6)
plt.xticks(np.arange(0,100,5),color='orange')
plt.show()
```

```python
bims=np.arange(0,100,5)
plt.figure(figsize=(6,6))
plt.hist(marks,bins=bims,color='green',histtype='step')
plt.xticks(np.arange(0,100,5),color='orange')
plt.show()
```

Two plots in same graph

```python
marks=np.random.randint(0,100,(50))
marks2=np.random.randint(0,100,(50))
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))
plt.hist([marks,marks2],bins=bins,color=['blue','red'])
plt.xticks(np.arange(0,100,5))
plt.xlabel("Marks")
plt.ylabel("marks2")
plt.title("comparing marks")
plt.show()
```
