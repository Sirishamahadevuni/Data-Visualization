### plot bar


A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent. A bar Plot shows comparisions among discrete categories.One axis of the plot shows the specific categoosries being compared and the other axis represents a measured value

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
subjects=['Maths','English','Science','Social Studies','computers']
marks=[89,90,45,78,99]
```

```python
plt.bar(subjects,marks)
plt.show()
```

```python
plt.bar(subjects,marks,color='green')
plt.show()
```

```python
colors=['red','blue','green','pink','orange']
plt.bar(subjects,marks,color=colors)
plt.show()
```

```python
plt.bar(subjects,marks,color=colors,width=0.6)
plt.show()
```

```python
plt.bar(subjects,marks,color=colors,edgecolor='black')
plt.show()
```

```python
plt.bar(subjects,marks,color=colors,edgecolor='black',linewidth=2,linestyle='--')
plt.show()
```

```python
plt.barh(subjects,marks,color=colors)
plt.show()
```

```python
subjects=["maths",'English','science','social Studies','Computers']
marks1=[89,90,45,78,99]
marks2=[78,56,34,90,12]
```

```python
plt.figure(figsize=(8,5))

plt.bar(subjects,marks1)
plt.bar(subjects,marks2)

plt.xlabel("Subjects")
plt.ylabel("marks")
plt.show()
```

```python
subjects_len=np.arange(len(subjects))
width=0.4
```

```python
plt.figure(figsize=(8,8))

plt.bar(subjects_len,marks1,width=width)
plt.bar(subjects_len+width,marks2,width=width)

plt.xlabel("Subjects")
plt.ylabel("marks")
plt.show()
```

```python
plt.figure(figsize=(6,8))

plt.bar(subjects_len,marks1,width=width,color=colors)
plt.bar(subjects_len+width,marks2,width=width,color=colors,alpha=0.4)

plt.xlabel("Subjects")
plt.ylabel("marks")
plt.show()
```
